Scraping the tweets using your time line

```python
# To scrap the tweets from your timeline, you need to use a command -
# home_timeline()

home_tweets = connection.home_timeline(tweet_mode = "extended")

#home_tweets
```

```python
# Lets convert the above json output (which is a key value pair) in to a pandas data frame

# I will create a data frame which will have the following: 

# 1) date 2) text 3) user_name 4) favorite_count 5) retweet_count

# Step 1: Will create the empty list

date = []
text = []
user_name = []
favorite_count = []
retweet = []

# Step 2: use a for loop to iterate within home_tweets

for x in home_tweets:
    date.append(x.created_at)
    text.append(x.full_text)
    user_name.append(x.user.name)
    favorite_count.append(x.favorite_count)
    retweet.append(x.retweet_count)
    

# Step 3: To convert the empty list to pandas Data frame

df = pd.DataFrame({"Date" : date,
                  "Text" : text,
                  "User Name" : user_name,
                  "Favorite" : favorite_count,
                  "Retweet" : retweet})

df.head()
```

Scraping the tweets using a hashtag - this can be done using a command search()

```python
# Scrapping the tweets for the hashtag #tesla

tweets = connection.search(q = "#tesla", lang = 'en', 
                           tweet_mode = 'extended')
```

```python
# Lets convert this into a data frame

# Step 1: Will create the empty list

date = []
text = []
user_name = []
favorite_count = []
retweet = []
location = []

# Step 2: use a for loop to iterate within home_tweets

for x in tweets:
    date.append(x.created_at)
    text.append(x.full_text)
    user_name.append(x.user.name)
    favorite_count.append(x.favorite_count)
    retweet.append(x.retweet_count)
    location.append(x.user.location)
    

# Step 3: To convert the empty list to pandas Data frame

df = pd.DataFrame({"Date" : date,
                  "Text" : text,
                  "User Name" : user_name,
                  "Favorite" : favorite_count,
                  "Retweet" : retweet,
                  "Location" : location})

df.shape
```

How to scrape more number of tweets

```python
# Lets try to scrap 300 tweets from #ItalianGP

# Lets create a empty data frame

df = pd.DataFrame(columns = ['Date', 'Text', 'User_Name', 'Favorite',
                            'Retweet', 'Location'])

df
```

```python
# This is the most important part - Creating an streaming user define 
# function. This will be used to extract the tweets based on the numbers
# you want and will safe the tweets in an excel file

def stream(query, file_name):
    i = 0
    for x in tweepy.Cursor(connection.search, q = query, lang = 'en',
                              tweet_mode = 'extended').items():
        print(i, end ='\r')
        df.loc[i,'Date'] = x.created_at
        df.loc[i, 'Text'] = x.full_text
        df.loc[i, 'User_Name'] = x.user.name
        df.loc[i, 'Favorite'] = x.favorite_count
        df.loc[i, 'Retweet'] = x.retweet_count
        df.loc[i, 'Location'] = x.user.location
        df.to_excel('{}.xlsx'.format(file_name))
        i+=1
        if i == 300:
            break
        else:
            pass
```

```python
stream(query = ['#ItalianGP'], file_name = "italy")
```

The biggest problem here is that none of the method to analyize text, understand the data in text format. Thus we need to convert the text into its numerical form. The most important method to convert text into number is DTM - Document term matrix
