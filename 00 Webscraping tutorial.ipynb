- Write a function called get_page that takes job_title, city and province as parameters
- Build the url with these parameters
- Make a request to the url
- Convert it a soup object
- Return the soup object

```python
def get_page(job_title,city,province):
#     job_title = input('Job Title:')
#     city = input('City:')
#     province = input('Province:')
    url = 'https://ca.indeed.com/'
    new_url = url + 'jobs?q=' + job_title + '&l=' + city + ',+' + province
    result1 = requests.get(new_url)
    soup1 = BeautifulSoup(result1.content)
    return soup1
     
```

```python
def get_page(job_title,city,province,page):
    if page == 1:
        x = 0
    else:
        x = (page - 1) * 10 
    url = f'https://ca.indeed.com/jobs?q={job_title}&l={city},+{province}&start={x}'
    result1 = requests.get(url)
    soup1 = BeautifulSoup(result1.content)
    #return soup1.find('div',id='searchCountPages').get_text(strip = True)
    return soup1
```

```python
get_page('data+scientist', 'toronto', 'on', 1)
```

```python
numpages
```

```python
Companylist=  []
for page_num in range(1, (numpages // 10) + 1):
    soup_page = get_page('data+scientist', 'toronto', 'on', page_num)
    for tag in soup_page.find_all('span',class_='companyName'):
        list = tag.get_text(strip=True)
        Companylist.append(list)
    
Companylist
```

```python
df = pd.DataFrame(Companylist)
```

```python
df.rename(columns={0:"CompanyName"}, inplace = True)
```
